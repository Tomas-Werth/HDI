```python
import dill
import numpy as np
import pandas as pd
```

```python
args={"dir":"C:/Users/tomas/Desktop/Desafio-MLE/", "sep":"|", "data":"claims_dataset.csv"} #Se define args como una libreria con variables claves

df=pd.read_csv(args["dir"]+args["data"],sep = args["sep"]) #Se abre el archivo con los requests

df["antiguedad_vehiculo"]=df["antiguedad_vehiculo"].fillna(1)
df["tipo_poliza"]=df["tipo_poliza"].fillna(1)
df["taller"]=df["taller"].fillna(1)
df["partes_a_reparar"]=df["partes_a_reparar"].fillna(3)
df["partes_a_reemplazar"]=df["partes_a_reemplazar"].fillna(1)
#imputamos los datos iniciales, la idea es inputarlos al comienzo por si los pipelines se ven afectados por la ausencia de algun dato.


pipeline=dill.load(open(args["dir"]+"pipeline_1.pkl" , 'rb'))
df=pipeline(df)
pipeline=dill.load(open(args["dir"]+"pipeline_2.pkl" , 'rb'))
df=pipeline(df)
df["log_total_piezas"]=df["log_total_piezas"].fillna(1.4545)
pipeline=dill.load(open(args["dir"]+"pipeline_3.pkl" , 'rb'))
df=pipeline(df)
df["marca_vehiculo_encoded"]=df["marca_vehiculo_encoded"].fillna(0).astype("int64")
pipeline=dill.load(open(args["dir"]+"pipeline_4.pkl" , 'rb'))
df=pipeline(df)
df["valor_vehiculo"]=df["valor_vehiculo"].fillna(3560).astype("int64")
df["valor_por_pieza"]=df["valor_por_pieza"].fillna(150)
pipeline=dill.load(open(args["dir"]+"pipeline_5.pkl" , 'rb'))
df=pipeline(df)
#Aplicamos los pipelines e imputamos donde corresponda, podemos ver que solo estamos imputando en orden, sin optimizar ningun proceso.


index = np.where(df["tipo_poliza"].values == 4)[0]
#definimos los indices de los request con "tipo_poliza"=4 para no ejecutar el modulo predictivo en ellos, ya que el cliente solicito un trato especial.

data_modelo=df.drop(index)[["log_total_piezas","marca_vehiculo_encoded","valor_vehiculo", "valor_por_pieza", "antiguedad_vehiculo"]]
model=dill.load(open(args["dir"]+"linnear_regression.pkl" , 'rb'))
#Ejecutamos el modelo predictivo solo con las columnas y requests necesarios.

results=model.predict(data_modelo)
for i in index:
    results=np.insert(results,i,-1)
results
#agregamos los valores  de los casos especiales y mostramos los resultados de las prediciones.
```

# Resolucion con request independientes

Librerias necesarias

```python
import os
import time
import dill
import numpy as np
import pandas as pd
```

Abrimos el documento inicialmente y comenzamos a trabajar los requests de forma individual

```python
args={"dir":"C:/Users/tomas/Desktop/Desafio-MLE/", "sep":"|", "data":"claims_dataset.csv"}
df=pd.read_csv(args["dir"]+args["data"],sep = args["sep"])
```

Para eso iteramos en cada fila del dataframe e intentamos definir el algoritmo como un proceso iterable.

```python
start = time.time() #Usamos time, para medir el tiempo de ejecucion del codigo
for index, row in df.iterrows(): #Iteramos en las filas del dataframe
    data=pd.DataFrame(row).transpose()
    data = data.astype(dtype= {"claim_id":"int64","marca_vehiculo":"object","antiguedad_vehiculo":"int64", "tipo_poliza":"int64", "taller":"int64", "partes_a_reparar":"int64", "partes_a_reemplazar":"int64"})
    #dejamos cada fila como un dataframe con una unica fila y asignamos los tipos correspondientes de las columnas.
    if pd.isna(data["antiguedad_vehiculo"].values):
        data["antiguedad_vehiculo"]=1
    if pd.isna(data["tipo_poliza"].values):
        data["tipo_poliza"]=1
    if pd.isna(data["taller"].values):
        data["taller"]=1
    if pd.isna(data["partes_a_reparar"].values):
        data["partes_a_reparar"]=3
    if pd.isna(data["partes_a_reemplazar"].values):
        data["partes_a_reemplazar"]=1
    #inputamos los valores nulos o faltantes de los datos iniciales

    
    data=dill.load(open(args["dir"]+"pipeline_1.pkl" , 'rb'))(data)
    data=dill.load(open(args["dir"]+"pipeline_2.pkl" , 'rb'))(data)
    data["log_total_piezas"]=data["log_total_piezas"].fillna(1.4545)
    data=dill.load(open(args["dir"]+"pipeline_3.pkl" , 'rb'))(data)
    data["marca_vehiculo_encoded"]=data["marca_vehiculo_encoded"].fillna(0).astype("int64")
    data=dill.load(open(args["dir"]+"pipeline_4.pkl" , 'rb'))(data)
    data["valor_vehiculo"]=data["valor_vehiculo"].fillna(3560).astype("int64")
    data["valor_por_pieza"]=data["valor_por_pieza"].fillna(150)
    data=dill.load(open(args["dir"]+"pipeline_5.pkl" , 'rb'))(data)
    #Ejecutamos los pipelines y las imputaciones de igual manera que el codigo anterior.

    
    if data["tipo_poliza"].values[0]!=4:
        data_modelo=data[["log_total_piezas","marca_vehiculo_encoded","valor_vehiculo", "valor_por_pieza", "antiguedad_vehiculo"]]
        model=dill.load(open(args["dir"]+"linnear_regression.pkl" , 'rb'))
        results=model.predict(data_modelo)[0]
    else:
        results=-1
    #En caso de que el "tipo_poliza" sea distinto de 4, ejecutamos el modelo predictivo, en caso contrario, asignamos -1 como resultado por la  condicion del cliente
    data["Semanas_en_el_taller"]=results
    #Guardamos el resultado de la predicion en una nueva columna llamada "Semanas_en_el_taller".
    if os.path.exists(args["dir"]+"Log.csv"): #chequeamos la existencia de un archivo "Log.csv" para no sobreescribir el resultado del request
        result_df=pd.read_csv(args["dir"]+"Log.csv", sep = args["sep"])
        result_df=result_df._append(data, ignore_index=True)
        result_df.to_csv(args["dir"]+"Log.csv", index=False, sep = args["sep"])
        # En caso de existir el archivo, este se abre, se actualiza y vuelve a guardarse.
    else:
        data.to_csv(args["dir"]+"Log.csv", index=False, sep = args["sep"])
        # En caso de no existir, se crea el archivo con los datos obtenidos.


end = time.time()
print("It took", end - start, "seconds!") #se mide el tiempo de ejecucion del codigo.
```

# Resolucion con Procesos en Paralelo

```python
import os
import json
import dill
import time
import subprocess
import numpy as np
import pandas as pd


with open("args.json", "r") as file:
    args = json.load(file)
df=pd.read_csv(args["dir"]+args["data"],sep = args["sep"])
#Se utiliza un archivo "args.json" para almacenar y leer los argumentos necesarios y se lee el archivo.


start = time.time()
for index, row in df.iterrows(): #Comienza la iteracion por fila
    data=pd.DataFrame(row).transpose()
    data = data.astype(dtype= {"claim_id":"int64","marca_vehiculo":"object","antiguedad_vehiculo":"int64", "tipo_poliza":"int64", "taller":"int64", "partes_a_reparar":"int64", "partes_a_reemplazar":"int64"})  
    data.to_csv(args["dir"]+"Dataframe_1.csv", index=False, sep = args["sep"])
    #Se guarda un archivo csv con la informacion de la fila correspondiente.

    subprocess.Popen(['python', args["dir"]+'Pipeline_1_2.py'])
    subprocess.Popen(['python', args["dir"]+'Pipeline_3_5.py'])
    #Comienzan 2 procesos en paralelo, el primero lee el csv recien creado y le aplica los pipelines 1 y 2, el segundo aplica los pipelines 3 y 5.
    #Ambos procesos son independientes y generan un csv con los datos tratados.
    
    
    flag=True  
    while flag==True:
        if os.path.exists(args["dir"]+"Dataframe_2.csv") and os.path.exists(args["dir"]+"Dataframe_3.csv"):
            flag=False
    #se espera a que ambos subprocesos generen los archivos correspondientes para poder continuar con el codigo.

    Dataframe_2=pd.read_csv(args["dir"]+"Dataframe_2.csv",sep = args["sep"])
    Dataframe_3=pd.read_csv(args["dir"]+"Dataframe_3.csv",sep = args["sep"])
    data=Dataframe_2.merge(Dataframe_3, how='outer')
    #Abrimos ambos archivos y los conmbinamos en un unico dataframe con la informacion generada por los pipelines 1,2,3 y 5.
    #Como el pipeline 4 depende tanto del 2 como del 3, se decicio ejecutarlo al final para aprovechar el procesamiento en paralelo.
    
    data=dill.load(open(args["dir"]+"pipeline_4.pkl" , 'rb'))(data)
    data["valor_vehiculo"]=data["valor_vehiculo"].fillna(3560).astype("int64")
    data["valor_por_pieza"]=data["valor_por_pieza"].fillna(150)
    #Se ejecuta el ultimo pipeline y se inputan los valores correspondientes
    

    if data["tipo_poliza"].values[0]!=4:
        data_modelo=data[["log_total_piezas","marca_vehiculo_encoded","valor_vehiculo", "valor_por_pieza", "antiguedad_vehiculo"]]
        model=dill.load(open(args["dir"]+"linnear_regression.pkl" , 'rb'))
        results=model.predict(data_modelo)[0]
    else:
        results=-1
    data["Semanas_en_el_taller"]=results
    if os.path.exists(args["dir"]+"Log.csv"):
        result_df=pd.read_csv(args["dir"]+"Log.csv", sep = args["sep"])
        result_df=result_df._append(data, ignore_index=True)
        result_df.to_csv(args["dir"]+"Log.csv", index=False, sep = args["sep"])
    else:
        data.to_csv(args["dir"]+"Log.csv", index=False, sep = args["sep"])

    #se continua al igual que el codigo anterior, pero  se eliminan los dataframes auxiliares que se usaron para los procesos en paralelo
    os.remove(args["dir"]+"Dataframe_1.csv")
    os.remove(args["dir"]+"Dataframe_2.csv")
    os.remove(args["dir"]+"Dataframe_3.csv")
    
    
end = time.time()
print("It took", end - start, "seconds!")
```

Notamos que el tiempo total se redujo. El subproceso de los pipelines 3 y 5 es considerablemente mas lento que el de los pipelines 1 y 2, pero el pipeline 4 no podria ejecutarse antes de que terminen ambos.
